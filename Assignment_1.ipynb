```python
from PyPDF2 import PdfReader
```

```python
file_paths_foxnews =  "data/file1.pdf"
file_path_reuters = "data/file2.pdf"
```

```python
reader = PdfReader(file_paths_foxnews)
text_foxnews = ""
for page in reader.pages:
    text_foxnews += page.extract_text() + "\n"  
print(text_foxnews)

reader2 = PdfReader(file_path_reuters)
text_reuters = ""
for page in reader2.pages:
    text_reuters += page.extract_text() + "\n"  
print(text_reuters)


```

While browsing the text, some spacing problems were noted. Certain sections exhibit extra spaces between words, while others lack necessary spaces between them. Despite these issues, the overall reading of the articles was satisfactory.

```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

# Analyzing the sentiment of each placeholder text
sentiment_foxnews = sia.polarity_scores(text_foxnews)
sentiment_reuters = sia.polarity_scores(text_reuters)

(sentiment_foxnews, sentiment_reuters)
```

Both texts have been assigned scores indicating that the news content is predominantly neutral. When comparing the levels of positivity and negativity, the text from Fox News displays a greater positive score, whereas Reuters exhibits a higher negative score. After reviewing the articles, I would rate the score from Reuters as being superior to that of Fox News

```python
headline_foxnews = "Embattled Harvard University President Claudine Gay hit with more plagiarism charges"
headline_reuters = "Harvard president will submit dissertation edits after plagiarism allegations"

simulated_sentiment_foxnews = sia.polarity_scores(headline_foxnews)
simulated_sentiment_reuters = sia.polarity_scores(headline_reuters)

(simulated_sentiment_foxnews, simulated_sentiment_reuters)
```

The score indicates a tendency towards neutrality, with a slight lean towards the negative, and no indication of positivity. In this instance, I concur with the sentiment analysis

```python
from readability import Readability



r_foxnews = Readability(text_foxnews)
r_reuters = Readability(text_reuters)


flesch_foxnews = r_foxnews.flesch()
print(f"Flesch Reading Ease: {flesch_foxnews.score}, Grade Levels: {flesch_foxnews.grade_levels}")

flesch_reuters = r_reuters.flesch()
print(f"Flesch Reading Ease: {flesch_reuters.score}, Grade Levels: {flesch_reuters.grade_levels}")

fk_foxnews = r_foxnews.flesch_kincaid()
print(f"Flesch-Kincaid Grade Level: {fk_foxnews.score}")

fk_reuters = r_reuters.flesch_kincaid()
print(f"Flesch-Kincaid Grade Level: {fk_reuters.score}")

dc_foxnews = r_foxnews.dale_chall()
print(f"Dale-Chall Score: {dc_foxnews.score}, Grade Levels: {dc_foxnews.grade_levels}")

dc_reuters = r_reuters.dale_chall()
print(f"Dale-Chall Score: {dc_reuters.score}, Grade Levels: {dc_reuters.grade_levels}")

```

The Flesch Reading Ease, Flesch-Kincaid Grade Level, and Dale-Chall Score are readability metrics indicating the difficulty level of a text. The Flesch Reading Ease score of 21.4 suggests the text is very hard to read, suitable for college graduates due to its complex sentence and word structure. The Flesch-Kincaid Grade Level of 19.2 also points to the text being understandable by those with education beyond high school, indicating college-level comprehension is required. Lastly, the Dale-Chall Score of 12.78, which assesses word familiarity and complexity, confirms that the text is challenging, best suited for readers with advanced education. Together, these scores indicate the text is complex, aimed at highly educated readers.

Both Fox News and Reuters articles are rated as very challenging to read, targeting college graduate level readers. Fox News scores slightly higher in difficulty across the Flesch Reading Ease, Flesch-Kincaid Grade Level, and Dale-Chall Score, indicating its content is marginally more complex. However, the differences are minimal, suggesting both sources require a high level of reading proficiency.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def generate_wordcloud(text, title):
    wordcloud = WordCloud(width = 800, height = 800, 
                          background_color ='white', 
                          stopwords = None, 
                          min_font_size = 10).generate(text)
                     
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    plt.title(title)
    plt.show()

generate_wordcloud(text_foxnews, "Fox News Article Word Cloud")
generate_wordcloud(text_reuters, "Reuters Article Word Cloud")
```

The word clouds from Fox News and Reuters articles highlight similar themes of academic controversy, with emphasis on "Havard", plagiarism," "allegations," "Gay" and "President." Key terms like "research," "integrity," and specific names suggest the content relates to an academic integrity issue. The prominence of such words indicates a shared focus on a significant controversy, likely involving a high-profile individual or institution.
