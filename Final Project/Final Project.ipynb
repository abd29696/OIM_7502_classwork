```python
import pandas as pd
```

```python
excel_file_path_1 = 'Query_for_BABSON_Project-2.xlsx'
excel_file_path_2 = 'EQUIPMENT_by_ITEM_PkgBOM_with_Cycle_Times_per_CASE.xlsx'

# Load the Excel file
xl_1 = pd.ExcelFile(excel_file_path_1)
xl_2 = pd.ExcelFile(excel_file_path_2)

# Dictionary to store DataFrames for each sheet
dfs = {}

# Iterate through each sheet name
for sheet_name in xl_1.sheet_names:
    
    # Read each sheet into a DataFrame
    dfs[sheet_name] = xl_1.parse(sheet_name)
    
# Iterate through each sheet name
for sheet_name in xl_2.sheet_names:
    
    # Read each sheet into a DataFrame
    dfs[sheet_name] = xl_2.parse(sheet_name)
    
for sheet_name, df in dfs.items():
    print(f"DataFrame for {sheet_name}:")
    print(df)
```
